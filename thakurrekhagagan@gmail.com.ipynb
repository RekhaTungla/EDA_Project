## Runtimes of the most revenue generating movies?

````python
runtime_rev = Movie_df.groupby('runtime').agg('mean')
top_10_runtime_rev = runtime_rev.sort_values('revenue', ascending=False)[:10]
top_10_runtime_rev.reset_index(inplace=True)
runtime_rev.reset_index(inplace=True)
top_10_runtime_rev.head(3)
````

Since, the Runtime has a big range.

Let's divide these times into SHORT, MEDIUM and LONG, by virtue of, ease of study.

Duration (in mins) Type of the Movie

66 - 100 Short

100 - 123 Medium

123 - 187 Long

````python
runtime_bins = [65,100,123,188]            #Time Intervals
runtime_labels = ['Short', 'Medium', 'Long']
runtime_rev['Type'] = pd.cut(runtime_rev.runtime, runtime_bins, labels=runtime_labels)
Movie_df['Length_of_movie'] = pd.cut(Movie_df.runtime, runtime_bins, labels=runtime_labels)
runtime_rev.reset_index(inplace=True)
runtime_rev.head(3)
````

````python
plt.figure(figsize = (10,7))
fontdict={'fontsize': 18,
          'weight' : 'bold',
         'horizontalalignment': 'center'}
sns.set_style('ticks')
sns.scatterplot(data=runtime_rev, hue='Type', x='runtime', y='revenue', palette='dark')
plt.xlabel('Runtime', fontsize=16, fontweight='bold')
plt.ylabel('Revenue', fontsize=16, fontweight='bold')
plt.legend(scatterpoints=1,
           bbox_to_anchor=(1, 0.7), loc=2, borderaxespad=1.,
           ncol=1,
           fontsize=14)
plt.title('Scatter Plot of Runtime vs Revenue', y=1.02,fontweight='bold',fontsize=15)
````

An Interesting observation here is that the CORRELATION of Revenue with Runtime changes from mild to low as Runtime increases.

Short and Medium movies have mild correlation

Long movies have a low correlation
